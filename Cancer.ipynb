```python
#Dropping these columns that aren't useful
df=df.drop(columns=['Unnamed: 32','id'])
```

```python
#Number of diagnosis that is Maliganit or benign
display(df['diagnosis'].value_counts())
sns.countplot(df['diagnosis'])
```

```python
#Here we plot a boxplot that show the range of M is bigger than B 
sns.catplot(x='diagnosis',y='texture_mean',data=df,kind='box')
```

```python
#This desplay an strong relationship that when raduis mean increase the perimeter_mean
#There are an strong correlation that is 0.99
sns.scatterplot(x='radius_mean',y='perimeter_mean',data=df,hue='diagnosis',alpha=0.2 )
df[['radius_mean','perimeter_mean']].corr()
```

```python
#This desplay an moderate relationship .
#There are an strong correlation that is 0.67
sns.lmplot(x='radius_mean',y='concavity_mean',data=df,hue='diagnosis' )
df[['radius_mean','concavity_mean']].corr()
```

```python
#This desplay an low  association .
#There are an weak correlation that is 0.17
sns.lmplot(x='radius_mean',y='smoothness_mean',data=df,hue='diagnosis' )
df[['radius_mean','smoothness_mean']].corr()
```

```python
reduis_malignant=df['radius_mean'][df['diagnosis']=='M']
reduis_benign=df['radius_mean'][df['diagnosis']=='B']
```

```python
#Here we show the distribution  that we see from 5 to 18 is bengin
plt.figure(figsize=(10,10))
sns.distplot(reduis_malignant,label='Malignant')
sns.distplot(reduis_benign,label='benign')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.distplot(df['texture_mean'][df['diagnosis']=='M'],label='Malignant',kde=False)
sns.distplot(df['texture_mean'][df['diagnosis']=='B'],label='benign',kde=False)
plt.legend()
```

```python
#This desplay an strong relationship that when raduis_worst increase the area_worst
#There are an strong correlation that is 0.98
sns.scatterplot(x=df['radius_worst'],y=df['area_worst'],hue=df['diagnosis'])
df[['radius_worst','area_worst']].corr()
```

```python
#This desplay an moderate association  
#There are an moderate correlation that is 0.34
sns.scatterplot(x=df['texture_worst'],y=df['area_worst'],hue=df['diagnosis'])
df[['texture_worst','area_worst']].corr()
```
